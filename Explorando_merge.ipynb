Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”

```python
df_plataformas['rating'] = df_plataformas['rating'].replace(np.nan, 'G')
```

De haber fechas, deberán tener el formato AAAA-mm-dd

```python
df_plataformas['date_added'] = pd.to_datetime(df_plataformas['date_added']).dt.date
```

```python
df_plataformas.head()
```

Los campos de texto deberán estar en minúsculas, sin excepciones

```python
df_plataformas['type'] = df_plataformas['type'].str.lower()
df_plataformas['title'] = df_plataformas['title'].str.lower()
df_plataformas['director'] = df_plataformas['director'].str.lower()
df_plataformas['cast'] = df_plataformas['cast'].str.lower()
df_plataformas['country'] = df_plataformas['country'].str.lower()
df_plataformas['rating'] = df_plataformas['rating'].str.lower()
df_plataformas['duration'] = df_plataformas['duration'].str.lower()
df_plataformas['listed_in'] = df_plataformas['listed_in'].str.lower()
df_plataformas['description'] = df_plataformas['description'].str.lower()
```

```python
df_plataformas.head(2)
```

El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
# Regular Expressions (Regex)
#Extract capture groups in the regex pat as columns in a DataFrame.
# \d -- decimal digit [0-9] (some older regex utilities do not support \d, but they all support \w and \s)
# \s -- (lowercase s) matches a single whitespace character -- space, newline, return, tab, form [ \n\r\t\f].
# \w -- (lowercase w) matches a "word" character: a letter or digit or underbar [a-zA-Z0-9_].
# * -- 0 or more occurrences of the pattern to its left
# + -- 1 or more occurrences of the pattern to its left, e.g. 'i+' = one or more i's
# Fuente: https://developers.google.com/edu/python/regular-expressions
df_plataformas[['duration_int', 'duration_type']] = df_plataformas['duration'].str.extract('(\d+)\s*(\w+)')
```

```python
df_plataformas.head(2)
```

```python
# Borramos el campo duration
df_plataformas.drop(['duration'], axis = 'columns', inplace=True)
```
